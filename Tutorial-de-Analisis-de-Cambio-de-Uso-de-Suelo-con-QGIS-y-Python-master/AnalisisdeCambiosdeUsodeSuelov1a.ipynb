```python
import pandas as pd
%pylab inline
%ls
```

```python
USVTlaxcalaSerieII = pd.read_csv('Txt/USVTlaxcalaSerieIIWGS84_Text.csv', index_col=1)
USVTlaxcalaSerieIII = pd.read_csv('Txt/USVTlaxcalaSerieIIIWGS84_Text.csv', index_col=1)
USVTlaxcalaSerieIV = pd.read_csv('Txt/USVTlaxcalaSerieIVWGS84_Text.csv', index_col=1)
USVTlaxcalaSerieV = pd.read_csv('Txt/USVTlaxcalaSerieVWGS84_Text.csv', index_col=1)
```

```python
USVTlaxcalaSerieII 
```

```python
TiposdeSuelo = USVTlaxcalaSerieII[['Area_has']]
TiposdeSuelo = TiposdeSuelo.rename(columns = {'Area_has':'USVTlaxcalaSerieII'})
TiposdeSuelo['USVTlaxcalaSerieIII']=USVTlaxcalaSerieIII['Area_has']
TiposdeSuelo['USVTlaxcalaSerieIV']=USVTlaxcalaSerieIV['Area_has']
TiposdeSuelo['USVTlaxcalaSerieV']=USVTlaxcalaSerieV['Area_has']
TiposdeSuelo.head()
```

```python
TiposdeSuelo['Diferencia']=TiposdeSuelo['USVTlaxcalaSerieV']-TiposdeSuelo['USVTlaxcalaSerieII']
TiposdeSuelo.sort('Diferencia')
```

```python
TiposdeSuelo_Transp = TiposdeSuelo.transpose()
TiposdeSuelo_Transp
```

```python
TiposdeSuelo_Transp['PASTIZAL INDUCIDO'][0:4].plot()
TiposdeSuelo_Transp['AGRICULTURA DE TEMPORAL ANUAL Y PERMANENTE'][0:4].plot()
TiposdeSuelo_Transp['AGRICULTURA DE RIEGO ANUAL'][0:4].plot()
TiposdeSuelo_Transp['BOSQUE DE PINO'][0:4].plot()
xticks(rotation = 'vertical')
figsize(12,6)
ylim(8e3,2e5)
yscale('log')
legend()
```

```python
TiposdeSuelo_Transp['VEGETACIÓN SECUNDARIA ARBUSTIVA DE BOSQUE DE PINO-ENCINO'][0:4].plot(label='VEG SEC ARBUSTIVA')
TiposdeSuelo_Transp['VEGETACIÓN SECUNDARIA ARBÓREA DE BOSQUE DE PINO'][0:4].plot(label='VEG SEC ARBOREA')
TiposdeSuelo_Transp['VEGETACIÓN SECUNDARIA ARBUSTIVA DE BOSQUE DE ENCINO-PINO'][0:4].plot(label='VEG SEC BOSQUE')
TiposdeSuelo_Transp['AGRICULTURA DE TEMPORAL ANUAL'][0:4].plot(label='AGRICULTURA TEMP')
xticks(rotation = 'vertical')
figsize(3.3,6)
ylim(2e2,3e6)
yscale('log')
legend()
```
